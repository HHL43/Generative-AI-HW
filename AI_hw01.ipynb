### 用多個圓形跟星星來畫出類似美國隊長盾牌的圖形

```python
import numpy as np
import matplotlib.pyplot as plt

def draw_captain_america_shield():
    # 定義圓的半徑與顏色
    radii = [1.0, 0.8, 0.6, 0.4, 0.2]  # 由外到內的圓半徑
    colors = ['red', 'white', 'red', 'blue', 'white']  # 盾牌顏色

    fig, ax = plt.subplots(figsize=(6, 6))
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    ax.set_aspect('equal')

    # 繪製圓環
    for radius, color in zip(radii, colors):
        circle = plt.Circle((0, 0), radius, color=color, ec='black')
        ax.add_patch(circle)

    # 繪製中央的星星
    angles = np.linspace(0, 2 * np.pi, 11)
    star_radius = np.array([0.1, 0.04] * 5 + [0.1])  # 星星比例
    x = star_radius * np.cos(angles)
    y = star_radius * np.sin(angles)
    ax.fill(x, y, 'blue')

    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_frame_on(False)
    plt.show()

# 呼叫函式畫出美國隊長盾牌
draw_captain_america_shield()
```
