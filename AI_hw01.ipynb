## 以matplotlib畫出螺旋形狀，根據基本雙螺旋後嘗試變化出其他圖形，像是半透明顏色、三螺旋、黑色背景


```python
# 導入套件
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### 1.雙螺旋

```python
# 使用 linspace 產生 0 到 4π 之間的 1000 個等間隔數值，賦值給變數 t
t = np.linspace(0, 4 * np.pi, 1000)
x1 = t * np.cos(t)
y1 = t * np.sin(t)

x2 = -t * np.cos(t)
y2 = -t * np.sin(t)

plt.plot(x1, y1, color='blue', linewidth=2, label='Spiral 1')
plt.plot(x2, y2, color='green', linewidth=2, label='Spiral 2')
plt.title('Double Spiral')

# 設定 x 軸和 y 軸的比例相同，使螺旋線呈現正確的形狀
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()
```

#### 2.加入半透明顏色後去除線條

```python
t = np.linspace(0, 4 * np.pi, 1000)

# 加入透明填充
plt.fill_betweenx(y1, x1, alpha=0.2, color='blue')
plt.fill_betweenx(y2, x2, alpha=0.2, color='red')

plt.title('Double Spiral with Transparency')
plt.axis('equal')
plt.grid(True)

plt.show()
```

#### 3.變成三螺旋，需要調整螺旋的起始角度

```python

t = np.linspace(0, 4 * np.pi, 1000)

# 三個螺旋的 x, y 座標
x1 = t * np.cos(t)
y1 = t * np.sin(t)

x2 = t * np.cos(t + 2 * np.pi / 3)  # 第二條螺旋，旋轉 120°
y2 = t * np.sin(t + 2 * np.pi / 3)

x3 = t * np.cos(t - 2 * np.pi / 3)  # 第三條螺旋，旋轉 -120°
y3 = t * np.sin(t - 2 * np.pi / 3)

# 繪製三螺旋
plt.plot(x1, y1, color='blue', linewidth=2, label='Spiral 1')
plt.plot(x2, y2, color='red', linewidth=2, label='Spiral 2')
plt.plot(x3, y3, color='purple', linewidth=2, label='Spiral 3')

# 設定標題與圖例
plt.title('Triple Spiral')
plt.axis('equal')
plt.legend()
plt.grid(True)

# 顯示圖形
plt.show()
```

#### 4.改變背景

```python
plt.style.use('dark_background')  # 設定黑色背景
plt.figure(figsize=(6,6))  # 設定畫布大小

plt.plot(x1, y1, color='cyan', linewidth=2, label='Spiral 1')
plt.plot(x2, y2, color='magenta', linewidth=2, label='Spiral 2')
plt.plot(x3, y3, color='yellow', linewidth=2, label='Spiral 3')

plt.title('Triple Spiral on Dark Background', color='white')
plt.axis('equal')
plt.legend()
plt.grid(True, linestyle='--', color='gray')

plt.show()
```
