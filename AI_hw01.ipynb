### 產生兩個對稱的螺旋形狀，看起來有點像雙漩渦

```python
# 導入套件
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 使用 linspace 產生 0 到 4π 之間的 1000 個等間隔數值，賦值給變數 t
t = np.linspace(0, 4 * np.pi, 1000)
x1 = t * np.cos(t)
y1 = t * np.sin(t)
x2 = -t * np.cos(t)
y2 = -t * np.sin(t)

plt.plot(x1, y1, color='blue', linewidth=2, label='Spiral 1')
plt.plot(x2, y2, color='green', linewidth=2, label='Spiral 2')
plt.title('Double Spiral')

# 設定 x 軸和 y 軸的比例相同，使螺旋線呈現正確的形狀
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()
```
